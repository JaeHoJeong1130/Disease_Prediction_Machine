```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import torchvision
import torchvision.transforms as transfroms
from torchvision import datasets, transforms
```

```python
class CustomDataset(Dataset): 
  def __init__(self):
    self.x_data = x_train
    self.y_data = y_train

  # 총 데이터의 개수를 리턴
  def __len__(self): 
    return len(self.x_data)

  # 인덱스를 입력받아 그에 맵핑되는 입출력 데이터를 파이토치의 Tensor 형태로 리턴
  def __getitem__(self, idx): 
    x = torch.FloatTensor(self.x_data[idx])
    y = torch.FloatTensor(self.y_data[idx])
    return x, y
```

```python
dataset = CustomDataset()
train_dataloader = DataLoader(dataset, batch_size=1, shuffle=True)
```

```python
class ValidDataset(Dataset): 
  def __init__(self):
    self.x_data = x_test
    self.y_data = y_test

  # 총 데이터의 개수를 리턴
  def __len__(self): 
    return len(self.x_data)

  # 인덱스를 입력받아 그에 맵핑되는 입출력 데이터를 파이토치의 Tensor 형태로 리턴
  def __getitem__(self, idx): 
    x = torch.FloatTensor(self.x_data[idx])
    y = torch.FloatTensor(self.y_data[idx])
    return x, y
```

```python
valid_dataset = ValidDataset()
valid_dataloader = DataLoader(valid_dataset, batch_size=1, shuffle=False)
```

```python
class NeuralNet(nn.Module):
    def __init__(self, input_size,output_size):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(input_size, 32) 
        self.fc2 = nn.Linear(32,16)
        self.fc3 = nn.Linear(16,output_size)
        
    def forward(self, x):
        out_fc1 = F.relu(self.fc1(x))
        # out_ac1 = F.relu(self.activation(out_fc1))
        out_fc2 = F.relu(self.fc2(out_fc1))
        # out_ac2 = self.activation(out_fc2)
        out_fc3 = F.softmax(self.fc3(out_fc2))
        return out_fc3
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
model = NeuralNet(6,148) # x_train 차원 6, 병의 개수 148개
model = model.to(device) # Upload the model to GPU
```

```python
import torchsummary
torchsummary.summary(model, input_size=(1,6))
```

```python
batch_size = 1
```

```python
# Define Optimizer: Stochastic Gradient Descent method
loss_fn = nn.CrossEntropyLoss()
# optimizer = torch.optim.SGD(model.parameters(), lr=0.05)
optimizer = torch.optim.SGD (model.parameters(), lr=0.05, momentum=0.9) 
# optimizer = torch.optim.Adam(model.parameters(), lr=0.05)
```

```python
loss_fn
```

```python
# Define Loss function (Cross Entropy Loss here)

# Train the model
total_step = len(train_dataloader)
epochs = 100
for epoch in range(epochs):
    for i, (images, labels) in enumerate(train_dataloader):  # mini batch for loop
        
        # Upload to gpu
        images = images.to('cuda')
        labels = labels.to('cuda')
        
        # print (images.shape)
        
        # Forward pass
        outputs = model(images)         # forward: Prediction
        # labels = labels.squeeze(dim=-1)
        loss = loss_fn(outputs, labels) # Calculate the loss with the predicted labels (outputs) & ground-truth labels (labels)
        
        # Backward pass & Optimize
        optimizer.zero_grad()
        loss.backward()  # Automatic gradient calculation (autograd)
        optimizer.step() # Update model parameter with requires_grad=True 
        
        if (i+1) % 100 == 0:
            print ('Epoch [{}/{}], Step [{}/{}], Loss: {:.4f}' 
                   .format(epoch+1, epochs, i+1, total_step, loss.item()))
    
```

```python
with torch.no_grad():
    correct = 0
    total = 0
    for images, labels in valid_dataloader:
        images = images.to('cuda')
        labels = labels.to('cuda')
        
        outputs = model(images)
        
        _, predicted = torch.max(outputs.data, 1)  # classificatoin model -> get the label prediction of top 1 
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    print('Accuracy of the Logistic Regression on Dataset: {} %'.format(100 * correct / total))
```
